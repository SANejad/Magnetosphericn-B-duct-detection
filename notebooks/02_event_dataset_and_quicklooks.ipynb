## Event Dataset Construction and Quicklook Visualization

This notebook exports one event-centered NetCDF file per interval from a CSV list and then generates Bt-over-EPSD quicklook plots for visual inspection.

```
from pathlib import Path
import sys

for candidate in [Path.cwd() / "src", Path.cwd().parent / "src"]:
    if candidate.exists() and str(candidate) not in sys.path:
        sys.path.append(str(candidate))
```

```
from duct_detection import EventConfig, export_event_datasets_from_csv, batch_plot_event_quicklooks
```

```
interval_csv = "time_interval_formatted_2016_3_1_to_7.csv"
sample_output_dir = "samples"
plot_output_dir = "plotted_events"
config = EventConfig(probes=(1,), output_folder=sample_output_dir)
```

```
exported_files = export_event_datasets_from_csv(
    interval_csv=interval_csv,
    probes=config.probes,
    config=config,
)
exported_files[:5] if exported_files else []
```
