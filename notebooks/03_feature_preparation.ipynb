## Feature Preparation for Machine Learning

This notebook converts event-centered NetCDF samples into ML-ready arrays for both tabular models (Random Forest) and image-based models (CNN/CNN-LSTM).

```
from pathlib import Path
import sys

for candidate in [Path.cwd() / "src", Path.cwd().parent / "src"]:
    if candidate.exists() and str(candidate) not in sys.path:
        sys.path.append(str(candidate))
```

```
from pathlib import Path
from duct_detection import build_rf_training_arrays, build_cnn_tensor_stack, load_label_map, save_ml_bundle
```

```
sample_folder = "samples"
labels_csv = Path("labels/example_event_labels.csv")  # replace with your curated label file
label_map = load_label_map(labels_csv if labels_csv.exists() else None)
```

```
feature_names, X_rf, y_rf, rf_metadata = build_rf_training_arrays(sample_folder=sample_folder, labels=label_map)
X_cnn, y_cnn, cnn_metadata = build_cnn_tensor_stack(sample_folder=sample_folder, labels=label_map, out_shape=(128, 64))
```

```
save_ml_bundle("ml_data", "rf", X_rf, y_rf, rf_metadata, feature_names=feature_names)
save_ml_bundle("ml_data", "cnn", X_cnn, y_cnn, cnn_metadata)
X_rf.shape, X_cnn.shape
```
