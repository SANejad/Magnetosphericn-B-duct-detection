## Feature Preparation for Machine Learning

Convert event-centered MMS datasets into ML-ready inputs (features, labels, and spectrogram representations).

```python
import os
import xarray as xr
import numpy as np
```

```python
data_folder = "samples"
files = [f for f in os.listdir(data_folder) if f.endswith(".nc")]

datasets = []
for f in files:
    ds = xr.open_dataset(os.path.join(data_folder, f))
    datasets.append(ds)
```

```python
X = []
y = []

for ds in datasets:
    bt = ds["Bt"].values
    density = ds["Density"].values

    features = [
        np.mean(bt),
        np.std(bt),
        np.mean(density),
        np.std(density),
    ]

    X.append(features)

    # placeholder label (you must replace this later)
    y.append(1)

X = np.array(X)
y = np.array(y)
```

```python
os.makedirs("ml_data", exist_ok=True)

np.save("ml_data/X_features.npy", X)
np.save("ml_data/y_labels.npy", y)
```
